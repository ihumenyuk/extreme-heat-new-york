```python
from dotenv import load_dotenv
load_dotenv('Keys.env')  # take environment variables

import pandas as pd
data_frame_1=pd.read_excel('Department-of-Education-–-Air-Conditioning-Report-1-of-2.xlsx', sheet_name='AC Status', header=1)
data_frame_2=pd.read_excel('Department-of-Education-–-Air-Conditioning-Report-1-of-2.xlsx', sheet_name='Building-School Crosswalk', header=1)
data_frame_2
data_frame_1=data_frame_1.rename(columns={'Building ID': 'Building Code'})
df=data_frame_1.merge(data_frame_2, how='left', on='Building Code')
```

```python
data_frame_3=pd.read_csv('2019_-_2020_School_Locations_20250707.csv')
data_frame_3
new_df=pd.merge(df, data_frame_3[['location_code', 'NTA', 'NTA_Name']], left_on='Building Code', right_on='location_code')
new_df
```

```python
def bin_ac(ac_value):
    if ac_value < 0.5: 
        return "red"
    if ac_value >= 0.5 and ac_value < 0.75:
        return "green"
    if ac_value >= 0.75 and ac_value < 0.95:
        return "yellow"
    else:
        return "blue"
    
```

```python
df["Percentage with A/C"].apply(bin_ac)
```

```python
df['bin_ac']=df["Percentage with A/C"].apply(bin_ac)
df
```
